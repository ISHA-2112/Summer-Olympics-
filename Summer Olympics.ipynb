# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
info=pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(info['City'].unique())
```

### 2. Which sport is having the most number of Gold Medals so far? (Top 5)

```python
f=info[info['Medal']=='Gold'] #filter the dataframe which consist of only gold medals
f.groupby("Sport")['Medal'].count().sort_values( ascending=False).head().plot.bar(10,10)
```

### 3. Which sport is having the most number of medals so far? (Top 5)

```python
info.groupby("Sport")['Medal'].count().sort_values( ascending=False).head().plot.bar(10,10)
```

### 4. Which player has won the most number of medals? (Top 5)

```python
info.groupby("Athlete")['Medal'].count().sort_values( ascending=False).head().plot.bar(10,10)
```

### 5. Which player has won the most number Gold Medals of medals? (Top 5)

```python
t=info[info['Medal']=='Gold']
t.groupby("Athlete")['Medal'].count().sort_values( ascending=False).head().plot.bar(10,10)
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
f=info[info['Medal']=='Gold'] #filter the data which consist of only gold medals
y=f[f['Country']=='IND']      #filter the data which consist of only India
y=y.sort_values(by='Year' ,ascending=True).head(1) #sorting in ascending order with respect to year
print(int(y['Year']))
```

### 7. Which event is the most popular in terms on number of players? (Top 5)

```python
lst=[]
for event in info['Event'].unique():
    lst.append([event, len(info[info['Event']==event])]) #appending the frequency of each event to the list
lst=pd.DataFrame(lst, columns=['Event','Players'])
lst.sort_values(by='Players',ascending=False).head().head().plot(x='Event',y='Players',kind='bar',figsize=(7,7))
```

### 8. Which sport is having the most female Gold Medalists? (Top 5)

```python
h=info[info['Medal']=='Gold']
y=h[h['Gender']=='Women']
y.groupby("Sport")['Medal'].count().sort_values( ascending=False).head().plot.bar(10,10)
```
